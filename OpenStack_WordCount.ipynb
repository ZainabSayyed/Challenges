```python
import pandas as pd
from nltk.corpus import stopwords 
import nltk
from nltk.corpus import treebank
import nltk.stem 
from nltk import stem
import numpy as np
import string
import re
import random
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
filename = "C:\\Users\\zaina\\Documents\\DataVisulaization_MachineLearning\\Open_Stack\\buglist.xlsx"
df_launchpad=pd.read_excel(filename)

filename = "C:\\Users\\zaina\\Documents\\DataVisulaization_MachineLearning\\Open_Stack\\report.xlsx"
df_stackoverflow = pd.read_excel(filename)
stop_words=stopwords.words('english')
```

```python
df=pd.DataFrame({'fault_description':random.sample(df_launchpad.fault_description,500),'req':random.sample(df_launchpad.req,500)},dtype=str)
df_st=pd.DataFrame({'longtext':random.sample(df_stackoverflow.longtext,500),'title':random.sample(df_stackoverflow.title,500),'tags':random.sample(df_stackoverflow.tags,500)},dtype=str)
```

```python
#Remove non-ascii character to process data using NLTK
#NLTK coverts string into meaningful words called token, removes punctuations and special characters.
#Analyzing Fault_desc and Req field from data set
#write process data to csv file
```

```python
corpus=[]
for i in df.fault_description:
    for k in i:
        if isinstance(k, str):
            k = unicode(k, "utf-8") 
    corpus.append(str(i.encode("utf-8")))         
df['fault_corpus']=pd.Series(corpus)
df['fault_corpus']=pd.Series( str(i).lower()  if len(str(i)) >0 else '' for i in df.fault_corpus)
corpus=[]
for i in df.fault_corpus:
    data=(str(i))
    data=re.sub(r'[^\x00-\x7f]','',data)
    corpus.append(data)  
df.fault_corpus=pd.Series(corpus)
df["fault_corpus_final"] = df.apply(lambda row: nltk.word_tokenize(row["fault_corpus"].translate(None, string.punctuation)), axis=1)
df["fault_words"]=df['fault_corpus_final'].apply(lambda x: [item for item in x if item not in stop_words])


corpus=[]
for i in df.req:
    for k in i:
        if isinstance(k, str):
            k = unicode(k, "utf-8") 
    corpus.append(str(i.encode("utf-8")))
df['req_corpus']=pd.Series(corpus)
df['req_corpus']=pd.Series( str(i).lower()  if len(str(i)) >0 else '' for i in df.req_corpus)

corpus=[]
for i in df.req_corpus:
    data=(str(i))
    data=re.sub(r'[^\x00-\x7f]','',data)
    corpus.append(data)  
df.req_corpus=pd.Series(corpus)
df["req_corpus"] = df.apply(lambda row: nltk.word_tokenize(row["req_corpus"].translate(None, string.punctuation)), axis=1)
df["req_words"]=df['req_corpus'].apply(lambda x: [item for item in x if item not in stop_words])


corpus=[]
for i in df_st.title:
    for k in i:
        if isinstance(k, str):
            k = unicode(k, "utf-8") 
    corpus.append(str(i.encode("utf-8")))         
df_st['title_corpus']=pd.Series(corpus)
df_st['title_corpus']=pd.Series( str(i).lower()  if len(str(i)) >0 else '' for i in df_st.title_corpus)


corpus=[]
for i in df_st.title_corpus:
    data=(str(i))
    data=re.sub(r'[^\x00-\x7f]','',data[1:])
    corpus.append(data)  
df_st.title_corpus=pd.Series(corpus)
df_st["title_corpus_final"] = df_st.apply(lambda row: nltk.word_tokenize(row["title_corpus"].translate(None, string.punctuation)), axis=1)
df_st["title_corpus_words"]=df_st['title_corpus_final'].apply(lambda x: [item for item in x if item not in stop_words])


corpus=[]
for i in df_st.tags:
    for k in i:
        if isinstance(k, str):
            k = unicode(k, "utf-8") 
    corpus.append(str(i.encode("utf-8")))         
df_st['tags_corpus']=pd.Series(corpus)
df_st['tags_corpus']=pd.Series( str(i).lower()  if len(str(i)) >0 else '' for i in df_st.tags_corpus)

corpus=[]

for i in df_st.tags_corpus:
    if (str(i)=='[]' or str(i)=='nan'):
        continue
    i=(re.sub(r'[["]','',i))
    i=i.replace("u'",'')
    i=(re.sub(r'[]"]','',i))
    data=i.split(',')
    substring=[]
    for j in data:
        substring.append(str(j.strip("'")))
    corpus.append(substring)
df_st["tags_corpus_final"]=pd.Series(corpus)

'''
corpus=[]
for i in df_st.longtext:
    for k in i:
        if isinstance(k, str):
            k = unicode(k, "utf-8") 
            corpus.append(str(i.encode("utf-8")))         

df_st['longtext_corpus']=pd.Series(corpus)
df_st['longtext_corpus']=pd.Series( str(i).lower()  if len(str(i)) >0 else '' for i in df_st.longtext_corpus)

corpus=[]
for i in df_st.longtext_corpus:
    data=(str(i))
    data=re.sub(r'[^\x00-\x7f]','',data)
    corpus.append(data)  
df_st.longtext_corpus=pd.Series(corpus)
df_st["longtext_corpus_final"] = df_st.apply(lambda row: nltk.word_tokenize(row["longtext_corpus"].translate(None, string.punctuation)), axis=1)
df_st["longtext_corpus_words"]=df_st['longtext_corpus_final'].apply(lambda x: [item for item in x if item not in stop_words])
'''
```

```python
#Creat dictionary for FaultDesc_Corpus and Req_Corpus
```

```python
tags_words_count=dict()
for w in df_st.tags_corpus_final:
    data=str(w)
    for i in data:
        tags_words_count[i]=tags_words_count.get(i,0)+1
fault_words_count=dict()
for w in df.fault_words:
    for i in w:
        fault_words_count[i]=fault_words_count.get(i,0)+1
req_words_count=dict()
for w in df.req_words:
    for i in w:
        req_words_count[i]=req_words_count.get(i,0)+1
title_words_count=dict()
for w in df_st.title_corpus_words:
    for i in w:
        title_words_count[i]=title_words_count.get(i,0)+1
longtext_words_count=dict()
for w in df_st.longtext_corpus_words:
    for i in w:
        longtext_words_count[i]=longtext_words_count.get(i,0)+1
```

```python
#Sorted dictionary for FaultDesc_Corpus and Req_Corpus
```

```python
mylist=[]
for key, value in sorted(fault_words_count.iteritems(), key=lambda (k,v): (v,k), reverse=True):
    mylist.append([key,value])
for key, value in sorted(req_words_count.iteritems(), key=lambda (k,v): (v,k), reverse=True):
    mylist.append([key,value])
for key, value in sorted(longtext_words_count.iteritems(), key=lambda (k,v): (v,k), reverse=True):
    mylist.append([key,value])
for key, value in sorted(title_words_count.iteritems(), key=lambda (k,v): (v,k), reverse=True):
    mylist.append([key,value])
for key, value in sorted(tags_words_count.iteritems(), key=lambda (k,v): (v,k), reverse=True):
    mylist.append([key,value])
df_wordcount=pd.DataFrame(mylist, columns =['Key','Count'])
```

```python
#Write dictionary of FaultDesc_Corpus and Req_Corpus to csv
for i in df_wordcount.Key:
    i=PorterStemmer().stem(i)
    snowball_stemmer = SnowballStemmer("english")
    i=snowball_stemmer.stem(i)
    i=WordNetLemmatizer().lemmatize(i)
```
